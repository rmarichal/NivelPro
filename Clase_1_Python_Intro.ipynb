## 🧐 ¿Qué es un notebook?
Un notebook es como un cuaderno interactivo:
- Puedes escribir código y ejecutarlo celda por celda.
- Puedes escribir texto para explicar lo que haces.
- Puedes incluir imágenes, gráficos y ecuaciones.
- Puedes guardar y compartir tu trabajo fácilmente.

Para ejecutar una celda de código en Colab, puedes presionar `Shift + Enter`.
